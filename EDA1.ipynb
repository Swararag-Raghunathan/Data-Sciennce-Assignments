# EXPLORATORY DATA ANALYSIS ON A DATASET

### 1.	Data Cleaning and Preparation:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Cardiotocographic.csv')
df
```

```python
df.head()
```

```python
df.info() #Complete information about the dataset
```

```python
df.isnull() #check for missing values
```

```python
df.isnull().sum()
```

```python
# Mean Imputation in column LB
mean_lb=df['LB'].mean()
df['LB'].fillna(mean_lb,inplace=True)
df['LB'].isnull().sum()
```

```python
# Mean Imputation in column AC
mean_ac=df['AC'].mean()
df['AC'].fillna(mean_ac,inplace=True)
df['AC'].isnull().sum()
```

```python
# Mean Imputation in column DS
mean_ds=df['DS'].mean()
df['DS'].fillna(mean_ds,inplace=True)
df['DS'].isnull().sum()
```

```python
# Mean Imputation in column DP
mean_dp=df['DP'].mean()
df['DP'].fillna(mean_dp,inplace=True)
df['DP'].isnull().sum()
```

```python
# Mean Imputation in column MLTV
mean_mltv=df['MLTV'].mean()
df['MLTV'].fillna(mean_mltv,inplace=True)
df['MLTV'].isnull().sum()
```

```python
# Mean Imputation in column Width
mean_width=df['Width'].mean()
df['Width'].fillna(mean_width,inplace=True)
df['Width'].isnull().sum()
```

```python
# Mean Imputation in column Tendency
mean_tendency=df['Tendency'].mean()
df['Tendency'].fillna(mean_tendency,inplace=True)
df['Tendency'].isnull().sum()
```

```python
# Mean Imputation in column NSP
mean_nsp=df['NSP'].mean()
df['NSP'].fillna(mean_nsp,inplace=True)
df['NSP'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
# Detect and treat outliers
z_scores=np.abs(stats.zscore(df))
outliers=(z_scores<3).any(axis=1)
df_no_outliers=df[outliers]
```

```python
df.info()
```

### 2. Statistical Summary:

```python
print(df_no_outliers.describe())
```

```python
print("Variable with high standard deviation:")
print(df_no_outliers.describe().loc['std'][df_no_outliers.describe().loc['std']>10])
print('\n')
print('Variable with high skewness:')
print(df_no_outliers.skew()[df_no_outliers.skew()>1])
print('\n')
print('Variable with high kurtosis:')
print(df_no_outliers.kurtosis()[df_no_outliers.kurtosis()>3])
```

### 4. Pattern Recognition and Insights:


```python
# Identify any correlation between variables and discuss thier potential implications.
corr_matrix = df_no_outliers.corr()
print("correlation matrix:",corr_matrix)

# Identify highly correlated variables.
highly_correlated_vars = []
for i in range(len(corr_matrix.columns)):
  for j in range(i+1,len(corr_matrix.columns)):
    col1 = corr_matrix.columns[i]
    col2 = corr_matrix.columns[j]
    correlation = abs(corr_matrix.iloc[i,j])
    if correlation >0.6:
      highly_correlated_vars.append((col1,col2,correlation))

 # Print highly correlated variables
print("Highly correlated variables:")
if highly_correlated_vars:
  for var in highly_correlated_vars:
    print(f"{var[0]} and {var[1]} are highly correlated with correlation of {var[2]:2f}")
else:
  print("No highly correlated variables found")
```

```python
 # Look for trends or patterns over time if temporal data is available.
if 'data' in df_no_outliers.columns or 'time' in df_no_outliers.columns:
  temporal_cols = ['date' if 'date' in df_no_outliers.columns else 'time']
  for col in numerical_cols:
    plt.figure(figsize=(10,6))
    plt.plot(df_no_outliers[temporal_cols],df_no_outliers[col])
    plt.title(f'Trend of {col} over Time')
    plt.xlabel('Time')
    plt.ylabel(col)
    plt.show()
else:
  print("No temporal data available for trend analysis.")
```
